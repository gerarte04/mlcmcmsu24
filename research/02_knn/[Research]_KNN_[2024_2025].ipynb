```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

def plot_knn_bound(X, y, scaler=None, n_neighbors=10, xlim=(-15, 15), ylim=(-20, 20)):
    # step size in the mesh
    h = 0.05

    # Create color maps
    cmap_light = ListedColormap(['C0', 'orange', 'cyan', 'green'][:np.unique(y).shape[0]])
    cmap_bold = ['C0', 'orange', 'c', 'darkgreen'][:np.unique(y).shape[0]]

    x_min, x_max = xlim
    y_min, y_max = ylim
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    grid = np.c_[xx.ravel(), yy.ravel()]

    X_scaled = X # if scaler is None
    if scaler is not None:
        grid = scaler.transform(grid)
        X_scaled = scaler.transform(X)

    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, algorithm='brute')
    clf.fit(X_scaled, y)

    Z = clf.predict(grid)

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y,
                    palette=cmap_bold, alpha=1.0, edgecolor="black")
    plt.xlabel('x_1')
    plt.ylabel('x_2')
    plt.title('   {}-NN'.format(n_neighbors))
    plt.grid()
    plt.show()
```

   $1$-$NN$  $10$-$NN$

```python
plot_knn_bound(X, y, n_neighbors=1)
plot_knn_bound(X, y, n_neighbors=10)
```

```python
from sklearn.metrics import accuracy_score

parameters = {
    'n_neighbors': [i for i in range(1, 11)],
    'metrics': ['cosine'],
    'weights': ['uniform', 'distance'],
    'normalizers': [(count_vec, 'CountVectorizer'), (tf_idf, 'TfidfVectorizer')]
}
```

```python
from cross_val import kfold_split, knn_cv_score
from sklearn.neighbors import KNeighborsClassifier

folds = kfold_split(X_train.size, 3)
scores = knn_cv_score(X_train, y_train, parameters, accuracy_score, folds, KNeighborsClassifier)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier

pipe = Pipeline([
    ('normalizer', tf_idf := TfidfVectorizer(max_df=0.8, min_df=10, max_features=1000, stop_words='english')),
    ('classifier', KNeighborsClassifier(n_neighbors=1, metric='cosine', weights='distance'))
])

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
import pandas as pd

cat_train_index = np.where(y_train == 7)[0]
tfidf_fit = tf_idf.fit_transform(X_train[cat_train_index])
train_words_list = tf_idf.get_feature_names_out()
train_weighted = tfidf_fit.toarray().sum(axis=0)
s1 = pd.Series(train_weighted, index=train_words_list)

cat_test_index = np.where(y_test == 7)[0]
tfidf_fit = tf_idf.fit_transform(X_test[cat_test_index])
test_words_list = tf_idf.get_feature_names_out()
test_weighted = tfidf_fit.toarray().sum(axis=0)
s2 = pd.Series(test_weighted, index=test_words_list)

words_list = list(set(train_words_list) & set(test_words_list))
```

```python
s1.size, s2.size
```

```python
import matplotlib.pyplot as plt

def draw_bar(x, y, title):
    plt.figure(figsize=(10, 5))
    plt.plot(x, y)
    
    plt.title(title)
    plt.xlabel('words')
    plt.ylabel('tf-idf scores')
    plt.xticks([])
    
    plt.show()

draw_bar(words_list, s1[words_list], 'Train sample words distribution')
draw_bar(words_list, s2[words_list], 'Test sample words distribution')
```
