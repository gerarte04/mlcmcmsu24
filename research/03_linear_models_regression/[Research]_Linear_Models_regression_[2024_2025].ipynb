```python
from sklearn.linear_model import Ridge
```

```python
np.random.seed(1)
X = np.random.uniform(0, 1, 100)
Y = X * 0.5 + 0.1 + np.random.randn(100) * 0.1

X3 = np.hstack((X[:, None], 3 * X[:, None]))
Y3 = X3[:, 0] * 0.5 + 0.1 + np.random.randn(100) * 0.1
```

```python
w1 = []
w2 = []

alphas = [0.01, 0.1, 1, 10, 100, 1000]

for alpha in alphas:
    reg = Ridge(alpha=alpha)
    reg.fit(X3, Y3)
    w1.append(reg.coef_[0])
    w2.append(reg.coef_[1])

w1 = np.array(w1)
w2 = np.array(w2)

fig, axs = plt.subplots(figsize=(14, 7), ncols=2)
axs[0].plot(alphas, w1, label="w1")
axs[0].plot(alphas, w2, label="w2")
axs[0].set_xscale("log")
axs[0].set_title("    alpha")
axs[0].set_xlabel("alpha")
axs[0].set_ylabel(" ")
axs[0].legend()
axs[1].plot(alphas, w2 / w1, label=" w2  w1", linewidth=10)
axs[1].plot([0.01, 1000], [3, 3], label=" x2  x1", linestyle="--", linewidth=8)
axs[1].set_xscale("log")
axs[1].set_ylim(2,4)
axs[1].set_xlabel("alpha")
axs[1].set_ylabel(" ")
axs[1].set_title(" ")
axs[1].legend()
plt.show()
```

```python
from sklearn.linear_model import Lasso
```

```python
reg = Lasso(alpha=1., max_iter=1000, tol=1e-4)
reg.fit(X3, Y3)
print(",  alpha = 1.")
print("w1:", reg.coef_[0], "\tw2:", reg.coef_[1])
print()

reg = Lasso(alpha=0.1, max_iter=1000, tol=1e-4)
reg.fit(X3, Y3)
print(",  alpha = 0.1")
print("w1:", reg.coef_[0], "\tw2:", reg.coef_[1])
print()

reg = Lasso(alpha=0.01, max_iter=1000, tol=1e-4)
reg.fit(X3, Y3)
print(",  alpha = 0.01")
print("w1:", reg.coef_[0], "\tw2:", reg.coef_[1])
print()

reg = Lasso(alpha=0.001, max_iter=1000, tol=1e-4)
reg.fit(X3, Y3)
print(",  alpha = 0.001")
print("w1:", reg.coef_[0], "\tw2:", reg.coef_[1])
print()

reg = Lasso(alpha=0.0001, max_iter=1000, tol=1e-4)
reg.fit(X3, Y3)
print(",  alpha = 0.0001")
print("w1:", reg.coef_[0], "\tw2:", reg.coef_[1])
print()

reg = Lasso(alpha=0.00001, max_iter=1000, tol=1e-4)
reg.fit(X3, Y3)
print(",  alpha = 0.00001")
print("w1:", reg.coef_[0], "\tw2:", reg.coef_[1])
print()
```

```python
np.random.seed(228)

X = np.arange(-200, 200, 1) * 0.05
Xr = X + np.random.randn(400) * 0.1
yr = np.sin(X) + np.random.randn(400) * 0.1
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(Xr[:, None], yr)
```

```python
xl = [-12, 12]
yl = xl * model.coef_ + model.intercept_
sns.lineplot(x=xl, y=yl, color='red')
sns.scatterplot(x=Xr, y=yr)
plt.title("Sinusoid data")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['regression line'])
```
